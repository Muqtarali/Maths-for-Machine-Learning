```python
from sklearn.datasets import make_regression

import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)
```

```python
plt.figure(figsize=(9, 6))
plt.scatter(X, y, color="red", label="Data points", alpha=0.7)
```

```python
# 2. Initial parameters (starting far off to show convergence)
b = -10
m = 15
lr = 0.05
epochs = 80
N = X.shape[0]
for i in range(epochs):
    # Predictions with current m and b
    y_pred = m * X + b

    # Compute errors
    error = y - y_pred

    # Calculate gradients for both slope (m) and intercept (b)
    slope_m = (-2 / N) * np.sum(X * error)
    slope_b = (-2 / N) * np.sum(error)

    # Update both parameters simultaneously
    m = m - (lr * slope_m)
    b = b - (lr * slope_b)

    # Plot lines periodically (fading older lines to show movement)
    if i % 2 == 0:  # Plot every 2nd epoch to keep the plot clean
        alpha_val = (i + 1) / epochs
        plt.plot(X, y_pred, color="blue", alpha=alpha_val * 0.4)

```
