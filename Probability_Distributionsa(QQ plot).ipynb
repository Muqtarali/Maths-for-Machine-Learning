```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
sns.kdeplot(data=df['sepal_length'])
```

```python
temp = sorted(df['sepal_length'].tolist())
```

```python
y_quant=[]
for i in range(1,101):
  y_quant.append(np.percentile(temp,i))
```

```python
samples = np.random.normal(loc=0,scale=1,size=1000)
```

```python
sns.scatterplot(x=x_quant,y=y_quant)
```

```python
# using statsmodel

import statsmodels.api as sm
import matplotlib.pyplot as plt

# Create a QQ plot of the two sets of data
fig = sm.qqplot(df['sepal_length'], line='45', fit=True)

# Add a title and labels to the plot
plt.title('QQ Plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Parameters for the underlying normal distribution
s = 0.5 # sigma
scale = np.exp(0) # exp(mu)

# Generate x-values and y-values
x = np.linspace(0.01, 5, 200)
pdf = lognorm.pdf(x, s, scale=scale)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(x, pdf)
plt.title('Lognormal Distribution Probability Density Function')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Real data
data = np.random.pareto(a=2, size=1000)

# Fit Pareto Distribution
shape, loc, scale = stats.pareto.fit(data)

# Distribution Object
dist = stats.pareto(b=shape, loc=loc, scale=scale)

# X values
x = np.linspace(data.min(), data.max(), 100)

# PDF
pdf = dist.pdf(x)

# Plot
plt.hist(data, bins=30, density=True, alpha=0.6)
plt.plot(x, pdf, color="red", linewidth=2)

plt.show()
```
